```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# 스킨푸드 스킨케어 크롤링 1
skin_name = [] # 스킨케어 제품이름 담을 리스트
skin_price = [] # 스킨케어 가격 담을 리스트
skin_imgs = []  # 스킨케어 사진 담을 리스트
skin_info = []  # 스킨케어 상세정보 담을 리스트
skin_type = [] # 스킨케어 타입 담을 리스트
skin_ingredient = [] # 스킨케어 성분 담을 리스트

driver = wb.Chrome()
driver.get('https://www.theskinfood.com/shop/shopbrand.html?type=N&xcode=014&mcode=001')
search = driver.find_elements(By.CLASS_NAME, 'MS_prod_img_l')
# 시선집중 아이템 지우기

# 전체화면 안하면 에러남
for i in range(4, 24, 1):
    search = driver.find_elements(By.CLASS_NAME, 'MS_prod_img_l')
    print(f"{i-3}번째 진행중")
    search[i].click()
    skin_name.append(driver.find_element(By.CSS_SELECTOR, 'h3').text)
    skin_price.append(driver.find_element(By.ID, 'pricevalue').text)
    skin_imgs.append(driver.find_element(By.CLASS_NAME, 'goods_thumbs-img').get_attribute('src'))
    skin_type.append(driver.find_element(By.CSS_SELECTOR, '#detail_product_info > div:nth-child(2) > ul > li').text)
    try:
        skin_info.append(driver.find_element(By.CSS_SELECTOR, '#img_container > img:nth-child(3)').get_attribute('src'))
    except:
        skin_info.append(driver.find_element(By.CSS_SELECTOR,'#goods_view > div.set_preload.goods-detail-tab-area > div.goods-detail-tab-content > div.goods_description.goods-detail-tab-content__item > div.img_container > div > img').get_attribute('src'))
    skin_ingredient.append(driver.find_element(By.CSS_SELECTOR, '#detail_product_info > div:nth-child(7) > ul > li').text)
    print("제목 : ", skin_name[i-4])
    print("가격 : ", skin_price[i-4])
    print("사진url : ", skin_imgs[i-4])
    print("상세정보 url : ", skin_info[i-4])
    print("타입 : ", skin_type[i-4])
    print("성분 : ", skin_ingredient[i-4])
    
    driver.back()
    time.sleep(1)

print("제목 : ", skin_name)
print("가격 : ", skin_price)
print("사진url : ", skin_imgs)
print("상세정보 url : ", skin_info)
print("타입 : ", skin_type)
print("성분 : ", skin_ingredient)
```
